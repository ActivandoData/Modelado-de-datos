# Funciones de 1 variable

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,2*np.pi,100)
y=np.sin(x)
plt.plot(x,y,'m--')
plt.grid()
w = np.linspace(0,2*np.pi,100)
v = np.sin(-w)
plt.plot(w,v,'g--')
```
