## Gr√°ficado de varias variables

```python
u = np.linspace(-2,5,100)
v = np.linspace(-2,5,100)

[x,y] = np.meshgrid(u,v)
# Ecuaci√≥n
z = np.exp(-x**2-y**2)
# Creaci√≥n de la figura
fig = plt.figure()
# Tama√±o
fig.set_size_inches(10,6)

# Se indica que es en 3D
ax = fig.gca(projection='3d')

# Generaci√≥n de la superficie
p = ax.plot_surface(x,y,z,cmap=cm.coolwarm)

# Valor de los colores
fig.colorbar(p,ax=ax)
```

```python
u = np.linspace(-5,5,100)
v = np.linspace(-6,6,100)

[x,y] = np.meshgrid(u,v)
# Ecuaci√≥n
z = np.cos(x)*np.cos(y)*np.exp(-np.sqrt(x**2+y**2)/20)
# Creaci√≥n de la figura
fig = plt.figure()
# Tama√±o
fig.set_size_inches(10,6)

# Se indica que es en 3D
ax = fig.gca(projection='3d')

# Generaci√≥n de la superficie
p = ax.plot_surface(x,y,z,cmap=cm.coolwarm)

# Valor de los colores
fig.colorbar(p,ax=ax)
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')

# Datos para tres dimensiones
zline = np.linspace(0, 15, 1000)
xline = np.sin(zline)
yline = np.cos(zline)
ax.plot3D(xline, yline, zline, 'gray')

# Datos para tres dimensiones con scatter plots en 3D
zdata = 15 * np.random.random(100)
xdata = np.sin(zdata) + 0.1 * np.random.randn(100)
ydata = np.cos(zdata) + 0.1 * np.random.randn(100)
ax.scatter3D(xdata, ydata, zdata, c=zdata, cmap='Greens');
```

### Curvas polares

```python
theta = np.linspace(0,2*np.pi)
r = 5 + 50*theta

fig = plt.figure()
ax = fig.add_subplot(111, projection="polar")
ax.plot(theta,r)

plt.show()
```

```python
theta = np.linspace(-np.pi,np.pi,100)
r = 2-4*(np.cos(theta))

fig = plt.figure()
ax = fig.add_subplot(111, projection="polar")
ax.plot(theta,r)

plt.show()
```

```python
theta = np.linspace(-np.pi,np.pi,100)
r = np.cos(4*theta)

fig = plt.figure()
ax = fig.add_subplot(111, projection="polar")
ax.plot(theta,r)

plt.show()
```

```python
theta = np.linspace(0,8*np.pi,100)
r = np.sin((8/5)*np.cos(theta))

fig = plt.figure()
ax = fig.add_subplot(111, projection="polar")
ax.plot(theta,r)

plt.show()
```
