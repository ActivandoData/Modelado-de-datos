## Regresión líneal con statsmodel 3

```python
import pandas as pd
import statsmodels.formula.api as smf
```

```python
df = pd.read_csv('../datasets/USvideos.csv')
df.head(100)
```

```python
reg = smf.ols('vistas ~ likes + comentarios + dislikes', data=df)
res = reg.fit()
```

```python
res.summary()
```
