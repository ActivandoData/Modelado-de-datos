## Regresión líneal con statsmodel 2
https://www.statsmodels.org/stable/gettingstarted.html

```python
import pandas as pd
import statsmodels.formula.api as smf
```

```python
df = pd.read_csv('../datasets/cars.csv')
```

```python
df.head(100)
```

```python
reg = smf.ols('consumo ~ cilindraje + potencia + peso',data=df)
# Realiza un ajuste
res = reg.fit()
```

```python
res.summary()
```

```python
# Para ver valor por valor
res.rsquared
```
